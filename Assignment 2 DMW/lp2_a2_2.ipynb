```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
dt=load_iris()
tdata=np.zeros(shape=(0,2))
for i in dt.data:
    tdata=np.vstack([tdata,[i[2],i[3]]])
print(tdata[:10])
```

```python
ddg=sch.dendrogram(sch.linkage(tdata,"ward"))
plt.title('Dendrogram')
plt.xlabel('Data')
plt.ylabel('Similarity')
plt.show()
```

```python
model=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
pred=model.fit_predict(tdata)
```

```python
plt.scatter(tdata[pred==0,0],tdata[pred==0,1],color='red')
plt.scatter(tdata[pred==1,0],tdata[pred==1,1],color='blue')
plt.scatter(tdata[pred==2,0],tdata[pred==2,1],color='green')
```
